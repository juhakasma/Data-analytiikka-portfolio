```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('http://taanila.fi/happy.xlsx')
df.head()
```

```python
df.plot.scatter('Freedom', 'Happiness Score')
```

```python
df[['Happiness Score','Freedom','Economy (GDP per Capita)'
         ,'Health (Life Expectancy)','Dystopia Residual']].corr().style.format('{:.2f}')
```

```python
ax=df.boxplot('Happiness Score', by='Region')
plt.title('')
plt.suptitle('')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
```

```python
ax=df.boxplot('Happiness Score', by='Region',figsize=(30,15))
plt.title('')
plt.suptitle('')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
```

```python
sns.barplot(x='Region',y='Happiness Score', data=df)
sns.set(rc={'figure.figsize':(15,8)})
```

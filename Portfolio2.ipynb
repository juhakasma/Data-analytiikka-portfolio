```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
raw_data = {'Valtio':['Suomi', 'Ruotsi', 'Norja', 'Tanska', 'Islanti']
            , 'BKT(milj)':[251900, 538000, 398800, 324900,23900]
           , 'Väestö (milj)':[5.5, 10, 5.3, 5.8, 0.3 ]
           , 'Työllisyysaste': ['75%', '84%', '72%', '77%', '88%']}
df = pd.DataFrame(raw_data)
df
```

```python
df.index=df['Valtio']
df
```

```python
plt.style.use('ggplot')
ax = df.sort_values(by='BKT(milj)').plot.barh(title='BKT Valtioittain',
                                               legend=True)
ax.set(xlabel='BKT(milj)', ylabel='Valtio')
vals=ax.get_xticks('BKT(milj)')
```

```python
df.pivot_table(values='BKT(milj)', index='Valtio'
                                ).sort_values('BKT(milj)', ascending=False)
```

```python
df.pivot_table(values='BKT(milj)', index='Valtio'
                                ).sort_values('BKT(milj)').plot.barh()
```

```python
ax = df.sort_values(by='BKT(milj)').plot.barh(title='Bruttokansantuote', legend=False)
ax.set(xlabel='(BKT(milj)', ylabel='Valtiot')
vals=ax.get_xticks()
ax.set_xticklabels(['{:.0f}'.format(x/1000) for x in vals])
```

```python
bktyht = df['BKT(milj)'].sum()
df['%osuus kokonais BKT:sta']=df['BKT(milj)']/bktyht*100
df
```

```python
Väestöyht = df['Väestö (milj)'].sum()
df['%osuus kokonaisväestöstä']=df['Väestö (milj)']/Väestöyht*100
df
```

```python
bkt1= df['%osuus kokonais BKT:sta']
bkt2= df['%osuus kokonaisväestöstä']
df['Tuotteliaisuuskerroin']=bkt1/bkt2
df
```

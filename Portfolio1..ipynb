```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_excel('http://taanila.fi/data1.xlsx', sheet_name='Data')
df.head()
```

```python
#Datan valmistelu
df.columns
```

```python
df.columns=['nro', 'sukupuoli', 'ikä', 'perhesuhde', 'koulutus', 'palveluv', 'palkka',
       'johto', 'työtov', 'työymp', 'palkkat', 'työteht', 'työterv', 'lomaosa',
       'kuntosa', 'hieroja']
df.columns
```

```python
df.rename(columns = {'sukupuoli': 'sukup', 'perhesuhde': 'perhe'}, inplace = True)
df.head()
```

```python
for var in df:
    print (var,pd.unique(df[var]))
```

```python
df.rename(columns = {'palkka': 'palkkaluokka', 'palveluv':'palveluvuodet'}, inplace = True)
df.head()
```

```python
bins = [1500, 2500, 3500, 4500, 5500, 6500]
df['palkkaluokka']=pd.cut(df['palkkaluokka'], bins=bins)
bins = [0,5,10,15,20,25,30,40]
df['palveluvuodet']=pd.cut(df['palveluvuodet'], bins=bins)
df.head()
```

```python
df[['palkkaluokka','palveluvuodet']].sort_values(by='palveluvuodet',ascending=False).head()
```

```python
#Lukumäärien ja prosenttien laskenta
palkkaluokka = ['1500', '2500', '3500', '4500', '5500+']
palveluvuodet = ['0-4v','5-9v','10-14v','15-19v','20-24v','25-29v','30v+']
```

```python
df1 = pd.crosstab(df['palkkaluokka'],'n')
df1.index = palkkaluokka
df1.columns.name = 'palkkaluokka'
df1
```

```python
df2 = pd.crosstab(df['palveluvuodet'], df['palkkaluokka'],normalize = 'columns')
df2.index = palveluvuodet
df2.columns.name = ''
df2.style.format('{:.1%}')
```

```python
ax = df2.plot.barh()
vals = ax.get_xticks()
ax.set_xticklabels(['{:.1%}'.format(x) for x in vals])
```

```python
ax = df2.transpose().plot.barh(stacked = True)
vals = ax.get_xticks()
ax.set_xticklabels(['{:.0%}'.format(x) for x in vals])
```

<b>Vaakapylväissä on laskettu palveluvuosien jakautuminen palkkaluokkiin<b>
    Oliosta näemme, ettei palvelusvuodet suoraan korreloi palkan kanssa.
